* **Method-III : The algo**
    - For understanding : [THEORY](https://www.geeksforgeeks.org/dsa/sort-an-array-of-0s-1s-and-2s/)
    - Code Given below: 

```python
a = [1,0,2,1,0,0,2,1,2,2]
low = mid = 0
high = len(a) - 1

def swap(a, x, y):
    a[x], a[y] = a[y], a[x]

while mid <= high:
    if a[mid] == 0:
        swap(a, low, mid)
        low += 1
        mid += 1
    elif a[mid] == 1:
        mid += 1
    else:  # a[mid] == 2
        swap(a, mid, high)
        high -= 1

print(a)
```

# Time Complexity

* **O(n)** because there is one loop of while.

# Space Complexity

* Constant -> O(1) -> because we changing in existing list/array
