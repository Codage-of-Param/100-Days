**`index_col`** : 

- We set column as index instead of sr no.

- ***str or list of str***  values

- It is optional

- default: None 

```python
pd.read_sql_query("SELECT * FROM country",conn, index_col=['Code']) 
```

**`coerce_float`** 

- Some databases store numbers as Decimal type (not regular numbers). When you read this data, coerce_float decides what to do with them.

`coerce_float=True` (default): Converts Decimal → regular float numbers

`coerce_float=False`: Keeps them as Decimal objects

- Decimal values kept as Decimal objects

=> **When to use each?**

- Use True: For normal math operations (easier to work with)

- Use False: When you need exact precision (like money calculations where every cent matters)
