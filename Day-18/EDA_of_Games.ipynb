#### How to check the skewed graph

- compare the **mean, median, and mode**. 

- In a **right-skewed distribution**, the ***mean is typically greater than the median***, while in a **left-skewed distribution**, the ***mean is less than the median.***

- The median is less affected by extreme values than the mean, making it a robust measure for skewed data

```python
print(df['critic_score'].median())
print(df['critic_score'].mean())

# nearly same (Not skewed)
```

```python
df['critic_score'].skew() # Skewness value
```
