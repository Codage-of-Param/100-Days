#### (5) developer

```python
x = df['developer'].isna() # It has some null values

df[x]
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
null_counts = df[categorical_cols].isnull().sum()
```

```python
plt.figure(figsize=(8,6))
null_counts.plot(kind='bar', color='maroon')
plt.title("Null Values in Categorical Columns")
plt.xlabel("Categorical Columns")
plt.ylabel("Number of Nulls")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- bar does not shows in developer because it has very less NaN values compare to the release_date and last_date

```python
plt.figure(figsize=(12,6))
sns.heatmap(df[categorical_cols].isnull(), cbar=False, cmap="viridis")
plt.title("Heatmap of Null Values in Categorical Columns")
plt.show()
```
