# **Bivariate Analysis**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Numerical-Numerical column

a. You can plot graphs like scatterplot(regression plots), 2D histplot, 2D KDEplots

b. Check correlation coefficent to check linear relationship

```python
df = pd.read_csv("..\\Day-19\\Video+Game+Sales\\vgchartz-2024.csv")
df
```

```python
df['critic_score'].corr(df['total_sales'])
# Gets a single correlation value between two specified Series (columns)
```

```python
df[['critic_score','total_sales']].corr() # Returns a correlation matrix for all numeric columns
```

- **It means `critic_score` and `total_sales` are highly corelated.**

#### 2D Scatter plot between `critic_score` and `total_sales`

```python
sns.relplot(data=df, x='critic_score', y='total_sales', kind='scatter', hue='genre')
```

#### 2D KDE plot between `critic_score` and `total_sales` :

```python
sns.displot(data=df, x='critic_score',y='total_sales', kind='kde')
# Nearly at point 7 all lines are near to each other it shows the mean/median for this particular data 
```

#### 1D KDE plot between `critic_score` and `total_sales` :

```python
sns.displot(data=df, x='critic_score',hue='genre', kind='kde') #Univarite plot
```

#### Box plot between `critic_score` and `total_sales` :

```python
# Correct: use 'column' instead of 'columns'
df.boxplot(column=['critic_score'], by='genre')
plt.xticks(rotation='vertical')
plt.show()
```

# 2. Numerical-Categorical column

a. create visualizations that compare the distribution of the numerical data across different categories of the categorical data.

b. You can plot graphs like barplot, boxplot, kdeplot violinplot even scatterplots

#### Plot between `genre` and `critic_score`

```python
# barplot

sns.catplot(data=df, kind='bar', x='genre', y='critic_score', hue='genre', aspect=2)
plt.xticks(rotation='vertical')
plt.show()
```

```python
# boxplot

sns.catplot(data=df, kind='box', x='genre', y='critic_score', hue='genre', aspect=1.5)
plt.xticks(rotation='vertical')
plt.show()
```

```python
# 2D histogram

sns.displot(data=df, x='genre', y='critic_score', hue='genre', aspect=1.5)
plt.xticks(rotation='vertical')
plt.show()
```

```python
# violin plot

sns.catplot(data=df, kind='violin', x='genre', y='critic_score', hue='genre', aspect=1.5)
plt.xticks(rotation='vertical')
plt.show()
```

```python
# stripplot

sns.stripplot(data=df, x='genre', y='critic_score', jitter=False)
plt.xticks(rotation='vertical')
plt.show()
```
