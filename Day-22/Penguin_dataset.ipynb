# Numerical column

#### 1. bill_length_mm

```python
penguin['bill_length_mm'].isna().sum()
```

```python
penguin[penguin['bill_length_mm'].isna()]
```

```python
median = penguin['bill_length_mm'].median()
penguin['bill_length_mm'].fillna(median, inplace=True)

penguin
```

```python
print(penguin['bill_length_mm'].mean())
print(penguin['bill_length_mm'].median())

# Mean and median are almost same so we can say that it is normally distributed
```

```python
penguin['bill_length_mm'].plot(kind='kde') 
```

```python
penguin['bill_length_mm'].skew()
```

```python
penguin['bill_length_mm'].plot(kind='box') # No outliers
```

# Categorical column

```python
# species	island  sex
```

#### 1. species

```python
penguin['species'].value_counts()
```

```python
penguin['species'].isnull().sum() # no null values
```

```python
values_species = penguin['species'].value_counts()
values_species.plot(kind='bar')
plt.xticks(rotation='horizontal')
plt.show()
```

```python
values_species.plot(kind='pie', autopct='%1.2f%%')
```

```python
penguin['species'].isna().sum() # no null values
```

#### 2. island

```python
penguin['island'].value_counts()
```

```python
penguin['island'].isna().sum()
```

```python
values_island = penguin['island'].value_counts()
sns.catplot(data=values_island, kind='bar')
plt.xticks(rotation='horizontal')
plt.show()
```

```python
values_island.plot(kind='pie', autopct='%1.2f%%')
```
