# 4. OneHotEncoding with Top categories

```python
counts = cars['brand'].value_counts()
counts

# Most of brands has less occurance so we can add all those column in new column others
```

```python
# So we need to find threshold from which value to not consider it in individual column

cars['brand'].nunique()
# Total 32 different brands are there
```

```python
thresh = 100 # means less then value of 100 are consider in others
```

```python
replaced = counts[counts <= thresh].index
replaced
```

```python
pd.get_dummies(cars['brand'].replace(replaced, 'others'))
```
