- **ColumnTransformer(transformers=[], remainder='passthrough')** we pass list of transformers and remainder has 2 values passthrough and drop , passthrough means je column ma transform nathy karvanu tene emnem j rakhse and if i write drop then it removes those column jema transform nthy thavanu

    <t>(transf_name, transformer(), applying_columns)<br>

```python
category_owner = [['First Owner', 'Second Owner', 'Third Owner','Fourth & Above Owner', 'Test Drive Car']]
```

```python
# we do simple imputer on km_driven, ordinal on owner, ohe on fuel 

Transformer = ColumnTransformer(transformers=[
    ('transformer-1', SimpleImputer(), ['km_driven']),
    ('transformer-2', OrdinalEncoder(categories = category_owner), ['owner']),
    ('transformer-3', OneHotEncoder(drop='first'), ['fuel'])
    ], remainder='passthrough')
```

```python
Transformer.fit_transform(X_train).shape
```
