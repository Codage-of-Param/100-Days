```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import scipy.stats as stats # for QQ plot
```

```python
df = pd.read_csv("D:\\100-Days clonned\\Day-30\\50_Startups.csv")
df
```

```python
df.shape
```

### Before applying transformation

```python
X = df.drop(columns=['State','Profit'])
Y = df.iloc[:,-1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
x_train
```

```python
y_train
```

```python
# linear regression model
from sklearn.metrics import r2_score


lr = LinearRegression() # object

lr.fit(x_train,y_train)

y_pred = lr.predict(x_test)

r2_score(y_test, y_pred)
```

```python
for feat in x_train:
    
    plt.figure(figsize=(12,4))
    
    plt.subplot(1,2,1)
    sns.distplot(x_train[feat])
    plt.title(feat)
    
    plt.subplot(1,2,2)
    stats.probplot(x_train[feat], dist="norm", plot=plt)
    plt.title(feat)
    
    plt.show()
```

### After applying transformation

```python
from sklearn.pipeline import FunctionTransformer
```

```python
trf = FunctionTransformer(func=np.log1p) # log1p = log(1+p)
```

```python
trf
```

```python
x_train_transformed = trf.fit_transform(x_train)
x_test_transformed = trf.transform(x_test)
y_train_transformed = trf.fit_transform(y_train)
y_test_transformed = trf.transform(y_test)
```

```python
# logistic regression model
from sklearn.metrics import r2_score

lr2 = LinearRegression() # object

lr2.fit(x_train_transformed, y_train_transformed)

y_pred2 = lr2.predict(x_test_transformed)

(r2_score(y_test_transformed, y_pred2))
```

```python
for feat in x_train_transformed:
    
    plt.figure(figsize=(12,4))
    
    plt.subplot(1,2,1)
    sns.distplot(x_train_transformed[feat])
    plt.title(feat)
    
    plt.subplot(1,2,2)
    stats.probplot(x_train_transformed[feat], dist="norm", plot=plt)
    plt.title(feat)
    
    plt.show()
```
