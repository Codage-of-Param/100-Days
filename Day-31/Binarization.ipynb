```python
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import Binarizer
```

```python
df = pd.read_csv("C:\\Users\\pcoda\\OneDrive\\Dataset{Manually}\\Kaggle\\Titanic\\train.csv", usecols=['Age', 'Fare', 'SibSp', 'Parch', 'Survived'])
```

```python
df
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
df.head()
```

```python
# Sibling spouse and parent child are working as family so we can add new feature with help of SibSp and Parch

df['Family'] = df['Parch'] + df['SibSp']
```

```python
df
```

```python
df.drop(columns=['SibSp', 'Parch'], inplace=True)
```

```python
df
```

```python
X = df.drop(columns='Survived')
Y = df['Survived']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
X_train
```

```python
Y_train
```

```python
# Before binarization

clf = DecisionTreeClassifier()

clf.fit(X_train, Y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
(accuracy_score(Y_test , y_pred))*100
```

```python
np.mean(cross_val_score(DecisionTreeClassifier(),X,Y,cv=10,scoring='accuracy'))*100
```

```python
# we do binarization on family , for example If person comes with family then we put 1 otherwise 0
```

```python
trfs = ColumnTransformer([
    ('family_binarization', Binarizer(copy=False), ['Family']) # copy=False meaning it changes in original datset
])
trfs
```

```python
X_train_trfs = trfs.fit_transform(X_train)
X_test_trfs = trfs.transform(X_test)
```

```python
X_train_trfs
```

```python
clf2 = DecisionTreeClassifier()
clf2.fit(X_train_trfs,Y_train)
y_pred2 = clf2.predict(X_test_trfs)

accuracy_score(Y_test,y_pred2)*100
```

```python
X_trf = trfs.fit_transform(X)
np.mean(cross_val_score(DecisionTreeClassifier(),X_trf,Y,cv=10,scoring='accuracy'))
```
