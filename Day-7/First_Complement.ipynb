```python
# Number Complement

decimal = int(input("Enter Number : "))
binary = ""

if decimal == 0:
    binary = "0"
else:
    while decimal > 0:
        remainder = decimal % 2
        binary = str(remainder) + binary
        decimal = decimal // 2
print(binary)   
```

```python
comp = ""
i = 0
for i in range(len(binary)):
    if binary[i] == '1':
        comp += '0'
    else:
        comp += '1'
print(comp)
        
```
